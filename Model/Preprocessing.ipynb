## Generating the output feature

```python
status_counts = df['loan_status'].value_counts()
for status, count in status_counts.items():
    percentage = (count / len(df)) * 100
    print(f"{status}: {count:,} ({percentage:.2f}%)")
```

```python
# Create binary target variable for default prediction
# 1 = Default (Charged Off, Default, Late payments)
# 0 = Good (Fully Paid)
# EXCLUDE Current loans - we don't know their final outcome yet

def create_default_target(loan_status):
    if loan_status in ['Charged Off', 'Default', 'Late (31-120 days)', 'Late (16-30 days)']:
        return 1  # Default
    elif loan_status == 'Fully Paid':
        return 0  # Good loan
    else:
        return None  # Exclude (Current)

df['default_flag'] = df['loan_status'].apply(create_default_target)

print(f"Shape before removing unknown outcomes: {df.shape}")
df.dropna(subset=['default_flag'], inplace=True)
print(f"Shape after removing unknown outcomes: {df.shape}")

print("\nDefault flag distribution (final dataset):")
print(df['default_flag'].value_counts())
print(f"\nDefault rate: {df['default_flag'].mean():.4f} ({df['default_flag'].mean()*100:.2f}%)")
```

```python
df.drop(['loan_status'], axis=1, inplace=True)
```

```python
# Analyze Grade and Sub-grade distributions
print("=== LOAN GRADE DISTRIBUTION ===")
grade_counts = df["grade"].value_counts().sort_index()
total_loans = len(df)

for grade, count in grade_counts.items():
    percentage = (count / total_loans) * 100
    print(f"Grade {grade}: {count:,} loans ({percentage:.2f}%)")

print(f"\nTotal unique grades: {df['grade'].nunique()}")
```

## Visualization

```python
# Grade distribution and default rates visualization using Seaborn/Matplotlib
import matplotlib.ticker as mtick

# Grade distribution
grade_counts = df['grade'].value_counts().sort_index()
plt.figure(figsize=(8,5))
sns.barplot(x=grade_counts.index,
            y=grade_counts.values,
            hue=grade_counts.index,
            palette='YlGnBu',
            dodge=False,
            legend=False)
plt.title('Loan Grade Distribution')
plt.xlabel('Grade')
plt.ylabel('Number of Loans')
plt.tight_layout()
plt.show()
```

```python
grade_stats = df.groupby('grade')['default_flag'].agg(['count','mean']).reset_index()
grade_stats.columns = ['grade','total_loans','default_rate']

plt.figure(figsize=(8,5))
sns.barplot(
    x='grade',
    y='default_rate',
    hue='grade',
    data=grade_stats.sort_values('grade'),
    palette='Oranges'
)

plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter(1.0))
plt.title('Default Rate by Grade')
plt.xlabel('Grade')
plt.ylabel('Default Rate')
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap for selected numeric features
numeric_cols = df.select_dtypes(include='number').columns.tolist()
plt.figure(figsize=(16,10))
corr_df = df[numeric_cols].corr()
sns.heatmap(corr_df, annot=True, fmt='.2f', cmap='PuBuGn', center=0)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
#mapping Grades 
grade_map = {'A':0, 'B':1, 'C':2, 'D':3, 'E':3, 'F':3, 'G':3}
df['grade'] = df['grade'].map(grade_map).astype('int64')

df['default_flag'] = df['default_flag'].astype(int)
```

```python
#mapping terms
df['term'] = df['term'].str.strip()
term_map = {'36 months': 0, '60 months': 1}
df['term'] = df['term'].map(term_map).astype('Int64')

#preprocessing home ownership
df['home_ownership'] = df['home_ownership'].str.upper().replace({'NONE':'OTHER', 'ANY':'OTHER'})
keep = ['MORTGAGE', 'RENT', 'OWN']
df['home_ownership_grp'] = df['home_ownership'].where(df['home_ownership'].isin(keep), 'OTHER')
df = pd.get_dummies(df, columns=['home_ownership_grp'])
df.drop('home_ownership_grp_OTHER', axis=1, inplace=True)
df.drop('home_ownership', axis=1, inplace=True)

#preprocessing verification status
verification_map = {'Not Verified': 0, 'Verified': 1, 'Source Verified': 2}
df['verification_status'] = df['verification_status'].map(verification_map).astype('int64')

#preprocessing purpose
main_purposes = ['debt_consolidation', 'credit_card', 'house', 'car', 
                'home_improvement', 'small_business', 'major_purchase']
df['purpose_grp'] = df['purpose'].where(df['purpose'].isin(main_purposes), 'OTHER')
df = pd.get_dummies(df, columns=['purpose_grp'], drop_first=True)
df.drop('purpose', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler

# List of numeric columns to scale
numeric_cols = ['dti', 'annual_inc', 'loan_amnt', 'revol_util', 'delinq_2yrs', 
                'inq_last_6mths', 'open_acc', 'pub_rec', 'bc_util', 'revol_bal', 
                'total_acc', 'mo_sin_old_rev_tl_op', 'num_actv_bc_tl', 'percent_bc_gt_75']

scaler = StandardScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

import joblib
joblib.dump(scaler, 'scaler.pkl')
```
