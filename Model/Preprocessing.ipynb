```python
# Selected features for credit risk model
features = [
    'dti',
    'annual_inc',
    'loan_amnt',
    'revol_util',
    'delinq_2yrs',
    'inq_last_6mths',
    'open_acc',
    'pub_rec',
    'bc_util',
    'revol_bal',
    'earliest_cr_line',
    'total_acc',
    'home_ownership',
    'verification_status',
    'purpose',
    'term',
    'mo_sin_old_rev_tl_op',
    'num_actv_bc_tl',
    'percent_bc_gt_75',
    'grade',
    'sub_grade',
    'loan_status'  # Target variable
]
```

```python
df = pd.read_csv("../Data/reduced_null_LCData.csv")
```

```python
status_counts = df['loan_status'].value_counts()
for status, count in status_counts.items():
    percentage = (count / len(df)) * 100
    print(f"{status}: {count:,} ({percentage:.2f}%)")
```

```python
# Create binary target variable for default prediction
# 1 = Default (Charged Off, Default, Late payments)
# 0 = Good (Fully Paid)
# EXCLUDE Current loans - we don't know their final outcome yet

def create_default_target(loan_status):
    if loan_status in ['Charged Off', 'Default', 'Late (31-120 days)', 'Late (16-30 days)']:
        return 1  # Default
    elif loan_status == 'Fully Paid':
        return 0  # Good loan
    else:
        return None  # Exclude (Current)

df['default_flag'] = df['loan_status'].apply(create_default_target)

print(f"Shape before removing unknown outcomes: {df.shape}")
df.dropna(subset=['default_flag'], inplace=True)
print(f"Shape after removing unknown outcomes: {df.shape}")

print("\nDefault flag distribution (final dataset):")
print(df['default_flag'].value_counts())
print(f"\nDefault rate: {df['default_flag'].mean():.4f} ({df['default_flag'].mean()*100:.2f}%)")
```

```python
# Analyze Grade and Sub-grade distributions
print("=== LOAN GRADE DISTRIBUTION ===")
grade_counts = df["grade"].value_counts().sort_index()
total_loans = len(df)

for grade, count in grade_counts.items():
    percentage = (count / total_loans) * 100
    print(f"Grade {grade}: {count:,} loans ({percentage:.2f}%)")

print(f"\nTotal unique grades: {df['grade'].nunique()}")

print("\n" + "="*50)
print("=== LOAN SUB-GRADE DISTRIBUTION ===")
subgrade_counts = df["sub_grade"].value_counts().sort_index()

for subgrade, count in subgrade_counts.items():
    percentage = (count / total_loans) * 100
    print(f"Sub-grade {subgrade}: {count:,} loans ({percentage:.2f}%)")

print(f"\nTotal unique sub-grades: {df['sub_grade'].nunique()}")
```
